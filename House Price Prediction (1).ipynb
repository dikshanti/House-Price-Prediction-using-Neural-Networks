## Train Test Split

```python
X = housing.drop('price',axis=1).values
```

```python
y = housing['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

# There are 19 features so, 19 neurons 
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define Early Stopping
early_stopping = EarlyStopping(
    monitor='val_loss',  # Metric to monitor (e.g., validation loss)
    patience=10,         # Number of epochs with no improvement before stopping
    restore_best_weights=True  # Restore the best model weights
)

# Train the model with Early Stopping
history = model.fit(
    x=X_train,
    y=y_train,
    validation_data=(X_test, y_test),
    batch_size=128,
    epochs=400,
    callbacks=[early_stopping]  # Include EarlyStopping in callbacks
)
```

```python
# this way we can see loss in training as well as testing dataset using validation data
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

## Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
mean_squared_error(y_test,predictions)**0.5
```

```python
mean_absolute_error(y_test,predictions)
```

```python
# best score is 1
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
```
